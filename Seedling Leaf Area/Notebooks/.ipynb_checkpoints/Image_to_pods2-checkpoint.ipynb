```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage as si
from scipy import ndimage
from scipy.spatial import distance as dist
from xml.dom import minidom
import xml.etree.ElementTree as ET

import matlab.engine
eng = matlab.engine.start_matlab()
```

```python
def order_points(pts):
    
    xSorted = pts[np.argsort(pts[:, 0]), :]
    leftMost = xSorted[:2, :]
    rightMost = xSorted[2:, :]
    leftMost = leftMost[np.argsort(leftMost[:, 1]), :]
    (tl, bl) = leftMost
    D = dist.cdist(tl[np.newaxis], rightMost, "euclidean")[0]
    (br, tr) = rightMost[np.argsort(D)[::-1], :]
    
    return np.array([tl, bl, br, tr], dtype="float32")
```

```python
#defining transformation function necessary in following step
def get_transform(p, scale):
    if scale == True:
        s = 0.04
    else:
        s = 0
    
    
    w1 = np.sqrt(((p[0][0][0] - p[3][0][0]) ** 2) + ((p[0][0][1] - p[3][0][1]) ** 2))
    w2 = np.sqrt(((p[1][0][0] - p[2][0][0]) ** 2) + ((p[1][0][1] - p[2][0][1]) ** 2)) 
 
    h1 = np.sqrt(((p[0][0][0] - p[1][0][0]) ** 2) + ((p[0][0][1] - p[1][0][1]) ** 2))
    h2 = np.sqrt(((p[2][0][0] - p[3][0][0]) ** 2) + ((p[2][0][1] - p[3][0][1]) ** 2))
    
    p1 = np.array([p[0][0][0] - s*w1, p[0][0][1] - s*w2])
    p2 = np.array([p[1][0][0] - s*w1, p[1][0][1] + s*w2])
    p3 = np.array([p[2][0][0] + s*w1, p[2][0][1] + s*w2])    
    p4 = np.array([p[3][0][0] + s*w1, p[3][0][1] - s*w2])
                   
    wmax = max(int((1+s)*w1), int((1+s)*w2))
    hmax = max(int((1+s)*h1), int((1+s)*h2))
    
    
    i = np.float32([p1, p2, p3, p4])
    o = np.float32([[0, 0], [0, hmax - 1], [wmax - 1, hmax - 1], [wmax - 1, 0]])
                        
    M = cv2.getPerspectiveTransform(i, o)
    
    return M, hmax, wmax
```

```python
#importing the image
os.chdir('C:\\Users\\sjoerd.kra\\Documents\\GitHub\\Datascience\\Seedling Leaf Area\\Notebooks')

dataset = '\Keria_1904'
img1 = cv2.imread(os.getcwd()+'\Data'+dataset+'\TrayImages\Keria_03_1904.jpg')
```

```python
#projecting the image to a perfect birds eye view
if img1.shape[1] >= img1.shape[0]:
    img1 = np.rot90(img1, 1, axes = (0, 1))


gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
gray_equalised = clahe.apply(gray)

ret, thresh = cv2.threshold(gray_equalised, 50, 255, cv2.THRESH_OTSU)

blob = si.morphology.binary_opening(ndimage.binary_fill_holes(~thresh), np.ones((20,20))).astype(np.uint8)
blob = ndimage.binary_fill_holes(si.morphology.binary_closing(blob, np.ones((100,100))).astype(np.uint8), np.ones([20,20])).astype(np.uint8)
contours, hierarchy = cv2.findContours(blob, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[-2:]
poly = cv2.approxPolyDP(contours[0][:,0,:], 500, True)

poly2 = np.empty(poly.shape)
poly2[:,0,:] = order_points(poly[:,0,:])
```

```python
M, hmax, wmax = get_transform(poly, True)

img2 = cv2.warpPerspective(img1, M, (wmax, hmax), flags=cv2.INTER_LINEAR)
blob2 = cv2.warpPerspective(blob, M, (wmax, hmax), flags=cv2.INTER_LINEAR).astype(np.uint8)

plt.imshow(img2)
plt.show()

contours2, hierarchy = cv2.findContours(blob2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[-2:]

rect = cv2.minAreaRect(contours2[0])
box = cv2.boxPoints(rect)
box = np.int0(box)

box2 = np.empty(poly.shape)
box2[:,0,:] = box

box3 = np.empty(box2.shape)
box3[:,0,:] = order_points(box2[:,0,:])

M, hmax, wmax = get_transform(box3, False)
img3 = cv2.warpPerspective(img2, M, (wmax, hmax), flags=cv2.INTER_LINEAR)

plt.imshow(img3)
plt.show()
```

```python
#cutting out individueal pods
                        
dh = int(img2.shape[0]/31)
dw = int(img2.shape[1]/20)
r = int(0.065*img2.shape[1])

centra = np.empty([30,10])

for i in range(30):
    centra[i,:] = dh*i
    
for i in range(10):
    centra[:,i] = dw*i
    
height, width = img2[:,:,0].shape

pods = np.zeros([2*r, 2*r, 3, 150])
k = 0

for i in range(0, 5):
    for j in range(0, 15):
        mask = np.zeros((height,width), np.uint8)
        cv2.circle(mask,((4*i+3)*dw, (2*j+1)*dh),r, 1,thickness=-1)
        pod = (mask[:, :, None]*img2)[max(0, ((2*j+1)*dh-r)):min(height, ((2*j+1)*dh+r)), 
                                      max(0, ((4*i+3)*dw-r)):min(width, ((4*i+3)*dw+r)), :]
        
        pods[:pod.shape[0],:pod.shape[1],:,k] = pod

        k = k+1
        
for i in range(0, 5):
    for j in range(1, 16):
        mask = np.zeros((height,width), np.uint8)
        cv2.circle(mask,((4*i+1)*dw, 2*j*dh),r, 1,thickness=-1)
        pod = (mask[:, :, None]*img2)[max(0, (2*j*dh-r)):min(height, (2*j*dh+r)), 
                                      max(0, ((4*i+1)*dw-r)):min(width, ((4*i+1)*dw+r)), :]
        
        pods[:pod.shape[0],:pod.shape[1],:,k] = pod

        k = k+1

```

```python
#saving the individual pod images
for i in range(pods.shape[3]):
    plt.imsave(os.getcwd()+'\Data'+dataset+'\PodImages\Keria_1904_'+str(i)+'.jpeg', (pods[:,:,:,i]/255))
```

```python
#importing the labels
label = np.empty([2, pods.shape[3]])

for h in range(pods.shape[3]):
    tree = ET.parse(os.getcwd()+'\Data\PodImages\Keria_03_1904_'+str(h)+'.xml')
    root = tree.getroot()
    n = len(list(root.iter('object')))
    center = np.empty([2,n])

    for i in range(n):
        xmin = int(root[5+i][4][0].text)
        ymin = int(root[5+i][4][1].text)
        xmax = int(root[5+i][4][2].text)
        ymax = int(root[5+i][4][3].text)
    
        center[:,i] = np.array([(xmax-xmin)/2, (ymax-ymin)/2])

    d = np.empty([sum(range(n))])
    k = 0
    
    for i in range(n):
        for j in range(i+1,n):
            d[k] = np.sqrt((center[0,i]-center[0,j])**2 + (center[1,i]-center[1,j])**2)
            k = k+1
    if k==0:
        davg = 0
    else:
        davg = np.mean(d)
        
    label[:,h] = np.array([n, davg])

```
