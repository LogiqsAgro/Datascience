```python
import pandas as pd
import numpy as np
import openpyxl
import xml.etree.cElementTree as et
import xmltodict
```

```python
df = pd.read_csv('../Data/dfKeria_1904.csv')
df.columns = [c.replace(' ', '_') for c in df.columns]
df
```

```python
dfcomp = pd.DataFrame(columns = ['Tray', 'Plug_Number', 'Number_of_plants_true'])
for T in range(1,4):
    for i in range (1,31):
        for j in range (1,6):
            df2 = pd.read_xml('../Data/xmldata/Keria1904xml/Keria_0'+ str(T) + '_1904_Plug_' + str(i) + '-' + str(j) + '.xml')
            ddf2 = df2['name'].value_counts().get(['Basil'],pd.Series(data=0))
            dfcomp = dfcomp.append({"Tray" : T, "Plug_Number": str(i) + '-' + str(j),'Number_of_plants_true': ddf2[0]}, ignore_index=True)
dfcomp
```

```python
df2 = pd.read_xml('../Data/xmldata/Keria1904xml/Keria_01_1904_Plug_1-1.xml')
df2
df2['name'].value_counts().get(['Basil'],0).values[0]
```

```python
dfcompare = df.merge(dfcomp, on=["Tray","Plug_Number"], how = 'inner')
dfcompare
```

```python
Esum = 0
for i in range (120,150):
    Esum += abs(dfcompare.at[i,'Number_of_plants'] - dfcompare.at[i,'Number_of_plants_true'])
for i in range (270,300):
    Esum += abs(dfcompare.at[i,'Number_of_plants'] - dfcompare.at[i,'Number_of_plants_true'])
for i in range (420,450):
    Esum += abs(dfcompare.at[i,'Number_of_plants'] - dfcompare.at[i,'Number_of_plants_true'])

MAE = Esum/90
MAE    
```

```python
EMax = 0
for i in range (0,450):
    E = abs(dfcompare.at[i,'Number_of_plants'] - dfcompare.at[i,'Number_of_plants_true'])
    if E > EMax:
        EMax = E
EMax    
```
