```python
from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image, plot_prediction_grid
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
```

```python
model = core.Model.load('model_weights_2.pth',['Basil', 'Nothing'])
```

```python
df = pd.DataFrame(columns=['Tray', 'Plug Number', 'Number of plants', 'Average spread'])
df
```

```python
imagecount = 150
plugperrow = 5
thresh=0.6
Tray = 1
mylist = []
for x in range (0,3):
    image_column = 1
    for i in range (0,plugperrow): 
        image_row = 1
        for j in range (0,int(imagecount/plugperrow)): 
            image = utils.read_image('../Data/Images/Keria_0' + str(Tray) + '_1904/Keria_0' + str(Tray) + '_1904_Plug_' + str(image_row) + '-' + str(image_column) + '.jpg')
            predictions = model.predict(image)
            labels, boxes, scores = predictions
            filtered_indices=np.where(scores>thresh)
            filtered_scores=scores[filtered_indices]
            boxes=boxes[filtered_indices]
            boxes = np.array(boxes)
            boxes = boxes.astype(int)
            Numberofplants = len(boxes[:,0])
            if Numberofplants == 0:
                avgspread = 0
            else:
                centers = np.zeros((Numberofplants,2))
                for s in range (0,Numberofplants):
                    centers[s,0] = (boxes[s,0] + boxes[s,2])/2
                    centers[s,1] = (boxes[s,1] + boxes[s,3])/2                      
                spread = 0
                for s in range (0,Numberofplants):
                    dist = np.zeros(Numberofplants)
                    for r in range (0,Numberofplants):
                        dist[r] = np.sqrt((centers[s,0] - centers[r,0])**2 + (centers[s,1] - centers[r,1])**2)
                    spread = spread+sum(dist)
                avgspread = spread/(Numberofplants**2)
            df = df.append({'Tray': str(Tray), 'Plug Number': str(image_row) + '-' + str(image_column),'Number of plants':Numberofplants, 'Average spread' :avgspread}, ignore_index=True)
            image_row += 1
        image_column+=1
    Tray+=1

                
                
                
            
```
